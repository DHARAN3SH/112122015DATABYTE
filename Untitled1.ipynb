```python
data=df[['Rooms','Bathroom','Car','Landsize','YearBuilt','Lattitude','Longtitude','Price']]
```

```python
median_Price = math.floor(data.Price.median())
median_Price
median_Bathroom = math.floor(data.Bathroom.median())
median_Bathroom
median_Car = math.floor(data.Car.median())
median_Car
median_Landsize = math.floor(data.Landsize.median())
median_Landsize
median_YearBuilt = math.floor(data.YearBuilt.median())
median_YearBuilt
median_Lattitude = math.floor(data.Lattitude.median())
median_Lattitude
median_Longtitude = math.floor(data.Longtitude.median())
median_Longtitude
```

```python
data.Lattitude = data.Lattitude.fillna(median_Lattitude)
data.Longtitude = data.Longtitude.fillna(median_Longtitude)

data.YearBuilt = data.YearBuilt.fillna(median_YearBuilt)
data.Landsize = data.Landsize.fillna(median_Landsize)
data.Car = data.Car.fillna(median_Car)
data.Bathroom = data.Bathroom.fillna(median_Bathroom)
data.Price = data.Price.fillna(median_Price)
```

```python
data
```

```python
# Defining the columns that containing the numerical data
#nc=numerical data
nc = ['Rooms','Bathroom','Car','Landsize','YearBuilt','Lattitude','Longtitude']
# Calculate the Z-scores for each and every column
z_scores = np.abs((df[nc]-df[nc].mean())/df[nc].std())
# Defining the zscore threshold for detecting outliers
zscore_threshold = 3
# Remove rows with outliers based on the Z-scores
data_no_outliers = data[(z_scores < zscore_threshold).all(axis=1)]
print(data_no_outliers)
```

```python
X = data[['Rooms','Bathroom','Car','Landsize','YearBuilt','Lattitude','Longtitude']]  # Features
y = data['Price']  # Target
```

```python
# Add a column of ones to X for the bias term
X['bias'] = 1
```

```python
X = X.to_numpy()
y = y.to_numpy()
```

```python
# Splitting the data into test and training datasets
test_ratio = 0.2  # 20% of the data will be used for testing
split_index = int(len(df) * (1 - test_ratio))
```

```python
X_train = X[:split_index]
X_test = X[split_index:]
y_train = y[:split_index]
y_test = y[split_index:]
```

```python
# Initialize the coefficients/parameters
coefficients = np.zeros(X_train.shape[1])
```

```python
# Define the hypothesis function (regression equation)
def hypothesis_function(X, coefficients):
    return np.dot(X, coefficients)
```

```python
# Define the loss function (mean squared error)
def loss_function(y_predicted, y_actual):
    return np.mean((y_predicted - y_actual) ** 2)
```

```python
# Define the gradient descent algorithm
def gradient_descent(X, y, coefficients, learning_rate, iterations):
    for _ in range(iterations):
        y_predicted = hypothesis_function(X, coefficients)
        error = y_predicted - y
        gradient = np.dot(X.T, error) / len(X)
        coefficients -= learning_rate * gradient
    return coefficients
```

```python
# Train the regression model using gradient descent
learning_rate = 0.01
iterations = 1000
coefficients = gradient_descent(X_train, y_train, coefficients, learning_rate, iterations)
```

```python
# Make predictions on the testing data
y_test_predicted = hypothesis_function(X_test, coefficients)
```

```python
# Calculate the mean squared error (MSE) for testing
mse_test = loss_function(y_test_predicted, y_test)
```

```python
print("Coefficients:", coefficients)
```

```python
print("Testing Mean Squared Error:", mse_test)
```
